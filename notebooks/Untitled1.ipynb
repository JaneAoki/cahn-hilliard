```python
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

```python
X = np.random.rand(256, 256)
```

```python
plt.contourf(X);
```

```python
nx, ny = X.shape[-2:]
xi, yi = np.indices((nx, ny))
```

```python
r = np.sqrt((xi-0.5*nx)**2 + (yi-0.5*ny)**2).astype(np.int)
```

```python
plt.contourf(r, len(np.unique(r.ravel())));
```

```python
ind = []
for i in np.unique(r.ravel()):
    xj = xi[np.where(r == i)]
    yj = yi[np.where(r == i)]
    ind.append((xj, yj))
```

```python
S = []
for xj, yj in ind:
    S.append(X[xj,yj].mean())
```

```python
plt.plot(np.array(S))
```

```python
nS = np.bincount(r.ravel(), X.ravel())/np.bincount(r.ravel())
```

```python
plt.plot(nS)
```
